## Step 4: Clean-up of the dataset

### Removing irrelevant columns

1. Identifier columns: just hashes, have no analytical value

```python
id_cols = ['account_rk', 'client_rk', 'order_rk']
```

2. Almost empty columns (>90% rows are empty)

```python
mostly_empty_cols = [
    'cancel_dttm',              # 100% empty
    'call_contact_6m_flg',      # 100 empty
    'call_contact_3m_flg',      # 100% empty
    'call_contact_1m_flg',      # 100% empty
    'bounce_cd',                # 95% empty
    'free_cancel_booking_dttm', # 91% empty
    'good_email_address_flg',   # 98% empty
    'bad_email_address_flg',    # 98% empty
    'last_email_send_dt',       # 98% empty
]
```

3. Other Columns

```python
duplicate_cols = ['local_book_start_dttm'] # the same as the "book_start_dttm" column, but in the local time zone.
# No sense to keep them both

noisy_cols = ['birth_place'] # Place of birth: the text is too loose (city, region, country — all mixed up),
# difficult to use in analysis
```

```python
cols_to_drop = id_cols + mostly_empty_cols + duplicate_cols + noisy_cols
df.drop(columns=cols_to_drop, inplace=True)
```

```python
print(f"After deleting: {df.shape[1]} columns left")
```

4. Correcting numeric columns

```python
numeric_str_cols = [
    'promo_code_discount_amt',
    'loyalty_accrual_rub_amt',
    'nominal_price_eur_amt',
    'nominal_price_rub_amt',
    'order_item_cnt',
    'month_beginning_balance_rub',
    'monthly_income_amt',
    'children_cnt',
    'age',
]
 
for col in numeric_str_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(',', '.', regex=False)
        .pipe(pd.to_numeric, errors='coerce')
    )
 
print("Numeric columns converted to float.")
```

5. Dates conversion

```python
date_cols = [
    'party_first_order_dt',
    'party_first_order_type_dt',
    'created_dttm',
    'book_start_dttm',
    'book_end_dttm',
    'last_sms_dt',
    'last_session_dttm',
]
 
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
 
print("Date columns converted into datetime.")
```

### Cleaning rows

1. Removing rows without orders

```python
# More than 49000 rows have no information in the field order_type_cd, 
before = len(df)
df = df[df['order_type_cd'].notna()].copy()
print(f"The number of rows without an order deleted: {before - len(df):,}")
```

2. Removing rows with incomplete orders

```python
before = len(df)
df = df[df['order_status_cd'] == 'SUC'].copy()
print(f"Incomplete orders deleted: {before - len(df):,}")
print(f"Rows remain: {len(df):,}")
```

3. Deleting rows with omissions in key fields

```python
key_cols = ['nominal_price_rub_amt', 'age']
before = len(df)
df.dropna(subset=key_cols, inplace=True)
print(f"Deleted lines with omissions in key fields: {before - len(df):,}")
```

### Saving cleaned dataset

```python
df.to_csv('travel_clean.csv', index=False)
```
